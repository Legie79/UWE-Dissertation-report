# Synthetic Data Generation
The code below create synthetic data that consist of the variables Patient_ID, Age,, Gender, Feeling_Stressed, Feeling_Anxious, Worried_about_future, Feeling_bored, Feeling_lonely, Sleep_problems, Internet_addiction.

```python
import random
def generate_synthetic_dataset(num_rows=1000): #give the function name and assign number of data needed
    Patient_ID = [f'P_{i+1:04d}' for i in range(num_rows)] #create patient ID in P_xxxx format upto 1000 rows 
    Age = [random.randint(18, 25) for i in range(num_rows)] #create random ages between 18 and 25.  
    Gender = [random.choice(['Male', 'Female', 'Other']) for i in range(num_rows)] # Create random gender from 'Male', 'Female', 'Other' 
    Feeling_stressed =[random.randint(0,1) for i in range(num_rows)] #Create random binary values(0 or 1) for "Feeling_sressted"variable.
    Feeling_anxious = [random.randint(0,1) for i in range(num_rows)] #Create random binary values(0 or 1) for this variable
    Worried_about_future = [random.randint(0,1) for i in range(num_rows)] #Create random binary values(0 or 1) for this variable
    Feeling_bored = [random.randint(0,1) for i in range(num_rows)] #Create random binary values(0 or 1) for this variable
    Feeling_lonely = [random.randint(0,1) for i in range(num_rows)] #Create random binary values(0 or 1) for this variable
    Sleep_problems = [random.randint(0,1) for i in range(num_rows)] #Create random binary values(0 or 1) for this variable
    Internet_addiction = [random.randint(0,1) for i in range(num_rows)] #Create random binary values(0 or 1) for this variable
    
    # Create DataFrame
    df = pd.DataFrame({
        'Patient_ID': Patient_ID,
        'Age': Age,
        'Gender': Gender,
        'Feeling_stressed' : Feeling_stressed,
        'Feeling_anxious' : Feeling_anxious,
        'Worried_about_future' : Worried_about_future,
        'Feeling_bored' : Feeling_bored,
        'Feeling_lonely' : Feeling_lonely,
        'Sleep_problems' : Sleep_problems,
        'Internet_addiction' : Internet_addiction
    })
    
    return df

# Generate the dataset
df = generate_synthetic_dataset()
df.head()
```

# Generate frequent Depression Symptoms issues using apriori

```python
# Convert DataFrame to 
symptoms = [] #initialise symptoms list to store the symptoms for each row.
for _, row in df_ARA.iterrows(): #iterate over row in the df_ARA dataframe
    symptom = [] #initialise symptom list to store the column names where the value is 1 from df_ARA
    for column in df_ARA.columns: 
        if row[column] == 1:  #
            symptom.append(column) #add that column into the sympton list
    symptoms.append(symptom) # add that lists to the symptoms list 

print(symptoms) # print the result. There will be thousand list in this result. 
```

```python
from apyori import apriori #importing apriori from the apyori library to create association riles
#Apply apriori algorithm to the list of symptoms created above. 
rules = apriori(transactions=symptoms, #dataset to be analysed. this is the list created before called symptoms 
                min_support=0.2, #minimum support threshold means a rule must appear in at least 20% of the times
                min_length=2, #minimum length threshold means rules must have at least 2 symptoms co-occurring
                max_length=0) #no upper limit for symptoms co-occurring
results = list(rules) #convert the rules created by the apriori algorithm into a list.
```

```python
results
```

```python
def AA_result(results): #define function named 'AA_result' that will return the result of AA
    #initialise empty list to store different components of the association rules.
    item_sets = []
    supports = [] 
    lhs = []  # antecedent
    rhs = [] # consequent
    confidences = []
    lifts = []
    for result in results: # interate through each result in the list of results created before by the AA 
        for each_rule in result[2]: #interate through each rule within the result
            #append itemset, support and ordered_statistics such as confidence and lift to the list initialised above
            item_sets.append(tuple(result[0]))
            supports.append(result[1])
            lhs.append(tuple(each_rule[0]))
            rhs.append(tuple(each_rule[1]))
            confidences.append(each_rule[2])
            lifts.append(each_rule[3])
    #return those tuple into a list with components of association rules 
    return list(zip(item_sets, lhs, rhs, supports, confidences, lifts))
#convert the list to a dataframe called result_df.
results_df = pd.DataFrame(AA_result(results),
                          columns = ['Item Set', 'Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```

# Apriori algorithm 
#### combining all codes to one cell to create function called apriori_algo

```python
import pandas as pd
from apyori import apriori

# Function to apply Apriori algorithm on a subset of the data
def apriori_algo(subset_df):
    transactions = []
    for index, row in subset_df.iterrows():
        transaction = []
        for column in subset_df.columns:
            if row[column] == 1:
                transaction.append(column)
        transactions.append(transaction)
    
    rules = apriori(transactions=transactions, 
                    min_support=0.2, 
                    min_length=2, 
                    max_length=0)
    results = list(rules)
    
    def AA_result(results):
        item_sets = []
        supports = [] 
        lhs = []
        rhs = []
        confidences = []
        lifts = []
        for result in results:
            for subset in result[2]:
                item_sets.append(tuple(result[0]))
                supports.append(result[1])
                lhs.append(tuple(subset[0]))
                rhs.append(tuple(subset[1]))
                confidences.append(subset[2])
                lifts.append(subset[3])
        return list(zip(item_sets, lhs, rhs, supports, confidences, lifts))
    
    results_df = pd.DataFrame(AA_result(results),
                              columns = ['Item Set', 'Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
    return results_df

```

#### Analysing the co-occurring symptoms by age group 

```python
age_groups = df_ARA['Age_Group'].unique() #extract unique age group from the dataframe 

results_list = [] # initialise an empty list to store the apriori result for each age group.

for age_group in age_groups: #loop through each age group
    subset_df1 = df_ARA[df_ARA['Age_Group'] == age_group].drop(columns=['Age_Group']) #create a subset of the dataframe for the age group created above and remove the columns in the dataframe
    apriori_results_df = apriori_algo(subset_df1) #apply the AA to the subset_df1
    apriori_results_df['Age_Group'] = age_group #add a new age_group column back to the dataframe
    results_list.append(apriori_results_df) # append the result to the list results_list

# Combine all results into a single DataFrame
all_results_df = pd.concat(results_list, ignore_index=True)

# Print the combined results DataFrame
print(all_results_df)
```
